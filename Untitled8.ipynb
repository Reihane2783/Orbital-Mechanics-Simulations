```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 4 * np.pi**2  # Gravitational constant in AU^3 / yr^2 / solar mass
M = 1.0           # Mass of the Sun in solar masses
dt = 0.001        # Time step in years
num_steps = 5000  # Number of simulation steps

# Initial position and velocity of Earth (in AU and AU/year)
x = 1.0
y = 0.0
vx = 0.0
vy = 2 * np.pi  # Tangential velocity for circular orbit

# Lists to store Earth's trajectory
x_list = []
y_list = []

# Euler-Cromer integration loop
for _ in range(num_steps):
    r = np.sqrt(x**2 + y**2)
    ax = -G * M * x / r**3
    ay = -G * M * y / r**3

    # Update velocities
    vx += ax * dt
    vy += ay * dt

    # Update positions using updated velocities
    x += vx * dt
    y += vy * dt

    x_list.append(x)
    y_list.append(y)

# Plotting the orbit
plt.figure(figsize=(6,6))
plt.plot(x_list, y_list, label='Earth trajectory')
plt.plot(0, 0, 'yo', label='Sun')
plt.xlabel('x (AU)')
plt.ylabel('y (AU)')
plt.title('Earth Orbit Simulation (Euler-Cromer Method)')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 4 * np.pi**2      # AU^3 / yr^2 / solar mass
M = 1.0               # Mass of the Sun
dt = 0.001            # Time step (years)
num_steps = 5000      # Simulation steps

# Initial position (1 AU from Sun)
x, y = 1.0, 0.0

# Initial velocity: less than circular to create ellipse
vx, vy = 0.0, 5.5     # Circular would be ~6.28 (2), so this is elliptical

# Lists to store trajectory
x_list = []
y_list = []

for _ in range(num_steps):
    r = np.sqrt(x**2 + y**2)
    ax = -G * M * x / r**3
    ay = -G * M * y / r**3

    vx += ax * dt
    vy += ay * dt

    x += vx * dt
    y += vy * dt

    x_list.append(x)
    y_list.append(y)

# Plotting the elliptical orbit
plt.figure(figsize=(6,6))
plt.plot(x_list, y_list, label='Elliptical Orbit (n = 2)')
plt.plot(0, 0, 'yo', label='Sun')
plt.xlabel('x (AU)')
plt.ylabel('y (AU)')
plt.title('Elliptical Orbit with Inverse-Square Gravity')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 4 * np.pi**2      # AU^3 / yr^2 / solar mass
M = 1.0               # Mass of the Sun
dt = 0.001            # Time step (years)
num_steps = 5000      # Simulation steps

# Earth's orbital eccentricity
e = 0.0167
a = 1.0               # Semi-major axis (AU)
x0 = a * (1 - e)      # Perihelion distance
y0 = 0.0

# Initial velocity from vis-viva equation
r0 = x0
v0 = np.sqrt(G * M * (2/r0 - 1/a))
vx0 = 0.0
vy0 = v0

# Exponents to test
powers = [1.5, 2.0, 3]

# Plot setup
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Orbital Behavior for Different Gravitational Exponents", fontsize=16)

for ax, n in zip(axes, powers):
    x, y = x0, y0
    vx, vy = vx0, vy0
    x_list, y_list = [], []

    for _ in range(num_steps):
        r = np.sqrt(x**2 + y**2)
        # Acceleration  1 / r^(n+1)
        ax_g = -G * M * x / r**(n + 1)
        ay_g = -G * M * y / r**(n + 1)

        vx += ax_g * dt
        vy += ay_g * dt

        x += vx * dt
        y += vy * dt

        x_list.append(x)
        y_list.append(y)

    ax.plot(x_list, y_list, label=f'n = {n}')
    ax.plot(0, 0, 'yo', label='Sun')
    ax.set_title(f'Force  1/r^{n}')
    ax.set_xlabel('x (AU)')
    ax.set_ylabel('y (AU)')
    ax.axis('equal')
    ax.grid(True)
    ax.legend()

# Changing the exponent from 2 causes orbital instability
# Only n = 2 produces closed, stable elliptical orbits (Bertrand's theorem)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 4 * np.pi**2
M = 1.0
dt = 0.001
num_steps = 5000

# Initial conditions
x0, y0 = 0.47, 0.0
vx0, vy0 = 0.0, 8.2

# ---------- Euler-Cromer ----------
x_ec, y_ec = x0, y0
vx_ec, vy_ec = vx0, vy0
x_list_ec, y_list_ec = [], []

for _ in range(num_steps):
    r = np.sqrt(x_ec**2 + y_ec**2)
    ax = -G * M * x_ec / r**3
    ay = -G * M * y_ec / r**3

    vx_ec += ax * dt
    vy_ec += ay * dt
    x_ec += vx_ec * dt
    y_ec += vy_ec * dt

    x_list_ec.append(x_ec)
    y_list_ec.append(y_ec)

# ---------- Runge-Kutta 2 ----------
x_rk2, y_rk2 = x0, y0
vx_rk2, vy_rk2 = vx0, vy0
x_list_rk2, y_list_rk2 = [], []

def accel(x, y):
    r = np.sqrt(x**2 + y**2)
    ax = -G * M * x / r**3
    ay = -G * M * y / r**3
    return ax, ay

for _ in range(num_steps):
    ax1, ay1 = accel(x_rk2, y_rk2)
    vx_mid = vx_rk2 + ax1 * dt / 2
    vy_mid = vy_rk2 + ay1 * dt / 2
    x_mid = x_rk2 + vx_rk2 * dt / 2
    y_mid = y_rk2 + vy_rk2 * dt / 2

    ax2, ay2 = accel(x_mid, y_mid)
    vx_rk2 += ax2 * dt
    vy_rk2 += ay2 * dt
    x_rk2 += vx_mid * dt
    y_rk2 += vy_mid * dt

    x_list_rk2.append(x_rk2)
    y_list_rk2.append(y_rk2)

# ---------- Runge-Kutta 4 ----------
x_rk4, y_rk4 = x0, y0
vx_rk4, vy_rk4 = vx0, vy0
x_list_rk4, y_list_rk4 = [], []

for _ in range(num_steps):
    ax1, ay1 = accel(x_rk4, y_rk4)
    k1vx = ax1 * dt
    k1vy = ay1 * dt
    k1x = vx_rk4 * dt
    k1y = vy_rk4 * dt

    ax2, ay2 = accel(x_rk4 + 0.5*k1x, y_rk4 + 0.5*k1y)
    k2vx = ax2 * dt
    k2vy = ay2 * dt
    k2x = (vx_rk4 + 0.5*k1vx) * dt
    k2y = (vy_rk4 + 0.5*k1vy) * dt

    ax3, ay3 = accel(x_rk4 + 0.5*k2x, y_rk4 + 0.5*k2y)
    k3vx = ax3 * dt
    k3vy = ay3 * dt
    k3x = (vx_rk4 + 0.5*k2vx) * dt
    k3y = (vy_rk4 + 0.5*k2vy) * dt

    ax4, ay4 = accel(x_rk4 + k3x, y_rk4 + k3y)
    k4vx = ax4 * dt
    k4vy = ay4 * dt
    k4x = (vx_rk4 + k3vx) * dt
    k4y = (vy_rk4 + k3vy) * dt

    vx_rk4 += (k1vx + 2*k2vx + 2*k3vx + k4vx) / 6
    vy_rk4 += (k1vy + 2*k2vy + 2*k3vy + k4vy) / 6
    x_rk4 += (k1x + 2*k2x + 2*k3x + k4x) / 6
    y_rk4 += (k1y + 2*k2y + 2*k3y + k4y) / 6

    x_list_rk4.append(x_rk4)
    y_list_rk4.append(y_rk4)

# ---------- Plot ----------
plt.figure(figsize=(8,8))
plt.plot(x_list_ec, y_list_ec, label='Euler-Cromer (2nd order)', color='orange')
plt.plot(x_list_rk2, y_list_rk2, label='Runge-Kutta 2nd order', color='limegreen')
plt.plot(x_list_rk4, y_list_rk4, label='Runge-Kutta 4th order', color='deepskyblue')
plt.plot(0, 0, 'yo', label='Sun')
plt.xlabel('x (AU)')
plt.ylabel('y (AU)')
plt.title('Orbit Comparison: Euler-Cromer vs RK2 vs RK4')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
